```python
from google.colab import files 
  
  
uploaded = files.upload()
```

```python
import pandas as pd 
import io 
  
dados = pd.read_csv(io.BytesIO(uploaded['data.csv'])) 
print(dados)
```

```python
dados
```

```python
previsores = dados.filter(regex=r"_mean", axis=1)
```

```python
previsores.head(10)
```

```python
y = dados["diagnosis"].map({"M": 0, "B": 1})
print(y)
```

```python
# Escalonamento dos atributos
from sklearn.preprocessing import StandardScaler

escala_base = StandardScaler()
X_base = escala_base.fit_transform(previsores)
```

```python
from sklearn.model_selection import train_test_split

X_base_treinamento, X_base_teste, y_base_treinamento, y_base_teste = train_test_split(X_base, y, test_size=0.50, random_state=0)
```

```python
X_base_treinamento.shape, X_base_teste.shape, y_base_treinamento.shape, y_base_teste.shape
```

```python
import pickle

with open("base.pkl", mode="wb") as df:
    pickle.dump([X_base_treinamento, y_base_treinamento, X_base_teste, y_base_teste], df)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
rede_neural = MLPClassifier(max_iter=7000, verbose=True, tol=0.000000001, solver= "adam", activation= "relu", hidden_layer_sizes=(100,100), batch_size=100, random_state=1)
rede_neural.fit(X_base_treinamento, y_base_treinamento)
```

```python
rede_neural = MLPClassifier(verbose=True, tol=0.000000001, solver= "adam", activation= "relu", batch_size=100, random_state=1)
rede_neural.fit(X_base_treinamento, y_base_treinamento)
```

```python
previsao = rede_neural.predict(X_base_teste)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(f"A precisão do algoritmo foi de {accuracy_score(y_base_teste, previsao) * 100}%")
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(rede_neural)
cm.fit(X_base_treinamento, y_base_treinamento)
cm.score(X_base_teste, y_base_teste)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# - Redes neurais ----------> 93.3333
paramentros = {"activation": ["relu", "logistic", "tahn"],
               "solver": ["adam", "sgd", "lbfgs"],
               "batch_size": [10, 56, 100, 200, 350, 400, 450]}

grid_search = GridSearchCV(estimator = MLPClassifier(), param_grid=paramentros)
grid_search.fit(X_base, y)
rn_melhores_paramentros = grid_search.best_params_
rn_melhores_resultados = grid_search.best_score_
```

```python
print(f"Os melhores parâmetros foram {rn_melhores_paramentros}. Os melhores resultados foram {rn_melhores_resultados * 100}%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_base_teste, previsao))
```
